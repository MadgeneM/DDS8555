```python
# Load the Carseats data set in Python
# Load necessary libraries and the Carseats dataset from seaborn-compatible source

import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

from ISLP import load_data
from ISLP.models import (ModelSpec as MS, summarize, poly)
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Fit a multiple regression model to predict Sales using Price, Urban, and US.
```

```python
# Load the Carseats dataset from ISLR via seaborn
carseats_data = load_data('Carseats')
#df = carseats_data
carseats_data.columns
```

```python
# Variables: Sales (continuous), Price (continuous), Urban (Yes/No), US (Yes/No)

# Convert categorical variables
carseats_data['Urban'] = carseats_data['Urban'].astype('category')
carseats_data['US'] = carseats_data['US'].astype('category')
```

```python
# Fit the multiple regression model (a)
model_a = smf.ols('Sales ~ Price + Urban + US', data=carseats_data).fit()
model_a_summary = model_a.summary()
```

```python
# Return the summary of the model
model_a_summary.tables[1]
```

**Interpretation:** When Price = 0, Urban = No, US = No, the predicted sales are 13.04 units (note: Price = 0 is unrealistic, so the intercept is mainly a baseline). Being in an urban areas increases sales by about 0.022 units, which is not statistically significant. Being located in the US increases sales by about 1.20 units, and this is statistically significant (p < 0.01). For each $1 increase in price, sales descrease by about 0.055 units, and this effect is highly significant (p < 0.001).

```python
# Fit a smaller model with only significant predictors
# Fit the multiple regression model (e)
model_e = smf.ols('Sales ~ Price + US', data=carseats_data).fit()
model_e_summary = model_e.summary()
```

```python
# Return the summary of the model (e)
model_e_summary.tables[1]
```

```python
# Identify potential outliers or high leverage observations in model (e).
diagnostics_cleaned = model_e.get_influence().summary_frame()
```

```python
# Identify potential outliers (|standardized residual| > 2)
outliers = diagnostics_cleaned[diagnostics_cleaned["standard_resid"].abs() > 2]
outliers
```

Observations with standardized residuals beyond plus or minus 2 are considered potential outliers. Observations with standardized residuals less than -2 are index 258, 143, 209, 106, and 57. These values indicate unusually low Sales relative to what the model predicted. Furthermore, the Studentized residuals (adjusted for leverage) for these are also beyond plus or minus 2, confirming the concern.

```python
# Identify high leverage points (leverage > 2(p+1)/n), p=2 predictors, n ≈ 400
n = len(diagnostics_cleaned)
p = 2
leverage_threshold = 2 * (p + 1) / n
high_leverage = diagnostics_cleaned[diagnostics_cleaned["hat_diag"].astype(float) > leverage_threshold]
high_leverage
```
